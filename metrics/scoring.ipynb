```python
import pandas as pd
from sklearn.metrics import ndcg_score, dcg_score
import numpy as np
```

```python
ranks = pd.read_csv("data.csv")
```

```python
from math import log2
c = 3
arr = list(map(lambda x: log2(x + 1), [i for i in range(1, c + 1)]))

parr = np.array(arr)
for i in range(0, 10):
  pred = ranks.drop(["Запрос", "Avg Precision", "NDCG(10)", "NDCG(5)", "NDCG(3)"], axis=1).iloc[i].to_list()[:c]
  true = np.array(sorted(pred, reverse=True))[:c]
  print(sum(np.divide(pred, parr)) / sum(np.divide(true, parr)), end=' ')
  # print(sum(np.divide(true, parr)))
#   print(ndcg_score([true], [pred]))
```
